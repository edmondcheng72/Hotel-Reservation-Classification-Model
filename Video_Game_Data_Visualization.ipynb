Data Visualization using Video Game Dataset

```python
#We will be using pandas for loading the csv file into a dataframe
import pandas as pd

#matplotlib is a python library for basic graph plotting
import matplotlib.pyplot as plt

#seaborn is another python graphing library that is built upon matplotlib and allows you to optimize matplotlib's output
import seaborn as sns

#switching from matplotlib's default graphing style to that of seaborn's. Seaborn displays graph lines, matplotlib doesn't
sns.set(color_codes = True)

#removing warnings from the output
import warnings
warnings.filterwarnings("ignore")
```

Import the data from a CSV file into a pandas dataframe

```python
games = pd.read_csv('https://raw.githubusercontent.com/edmondcheng72/Video-Game-Data-Visualization/main/vgsales.csv')
```

Taking a peek at the dataset

```python
games.head()
```

Finding how many records we have for each genre

```python
games['Genre'].value_counts()
```

Generating Heatmap

```python
#set the size of the figure for the plot (width: 10, height: 5)
plt.figure(figsize=(10, 5))

#create a heatmap using Seaborn (sns) library to visualize the correlation matrix of the 'games' DataFrame
#'cmap' parameter sets the color map to "Blues" for the heatmap
#'annot=True' displays the correlation values on the heatmap
#'linewidth=3' sets the width of the lines between cells in the heatmap
sns.heatmap(games.corr(), cmap="Blues", annot=True, linewidth=3)

#display the heatmap
plt.show()
```

Generating Count Plot

```python
#set the size of the figure for the plot (width: 10, height: 5)
plt.figure(figsize=(10, 5))

#create a count plot using Seaborn (sns) library, counting occurrences of each 'Genre' in the 'games' DataFrame
sns.countplot(data=games, x="Genre")

#set the title of the count plot to 'Frequency of Different Genres of Games'
plt.title('Frequency of Different Genres of Games')

#set the style of the plot to 'darkgrid'
sns.set_style("darkgrid")

#adjust the layout to prevent overlapping elements in the plot
plt.tight_layout()

#display the count plot
plt.show()
```

Generating Line Plot

```python
#set the size of the figure for the plot (width: 10, height: 5)
plt.figure(figsize=(10, 5))

#create a line plot using Seaborn (sns) library, with 'Year' on the x-axis, 'Global_Sales' on the y-axis,
#and different lines for each 'Genre' of video games
sns.lineplot(data=games, x="Year", y="Global_Sales", hue="Genre")

#set the title of the line plot to 'Global Sales of Video Games Over Time'
plt.title('Global Sales of Video Games Over Time')

#display the line plot
plt.show()
```

Generate Bar Plot

```python
#set the size of the figure for the plot (width: 10, height: 5)
plt.figure(figsize=(10, 5))

#create a bar plot using Seaborn (sns) library, displaying the average 'Global_Sales' for each 'Genre' of video games
sns.barplot(data=games, x="Genre", y="Global_Sales")

#set the title of the bar plot to 'Global Sales of Different Genre of Video Games'
plt.title('Global Sales of Different Genre of Video Games')

#rotate the x-axis labels (genres) by 45 degrees for better readability
plt.xticks(rotation=45)

#display the bar plot
plt.show()
```

Generating Pie Chart based on Regional Sales


```python
#creating new dataframe based on regional sales
regional_sales = games[['NA_Sales', 'EU_Sales', 'JP_Sales']]

#aggregate the sale values vertically & reset the index
regional_sales = regional_sales.sum().reset_index()
#renaming columns
regional_sales = regional_sales.rename(columns={"index": "Region", 0: "Sale"})
regional_sales
```

```python
#extracting 'Region' & 'Sale' column data from the new dataframe to use as labels and data
labels = regional_sales['Region']
sizes = regional_sales['Sale']

#creating a new figure for the plot with a specific size (width: 10, height: 8)
plt.figure(figsize=(10, 8))

#plotting a pie chart with specified sizes, labels, percentage format, and starting angle
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

#setting the title of the pie chart to 'Video Game Sales By Region'
plt.title('Video Game Sales By Region')

#displaying the pie chart with the defined figure size and properties
plt.show()
```

Generating Donut Chart

```python
#create a white circle to be placed at the center of the pie chart
my_circle = plt.Circle((0,0), 0.7, color='white')

#plot a pie chart with specified sizes, labels, percentage format, and starting angle
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

#set the title of the pie chart to 'Video Game Sales By Region'
plt.title('Video Game Sales By Region')

#get the current figure (plot) and add the white circle as an artist to the plot
p = plt.gcf()
p.gca().add_artist(my_circle)

#display the plot with the pie chart and the white circle in the center
plt.show()
```

Generating Scatterplot

```python
#set the size of the figure for the plot (width: 10, height: 5)
plt.figure(figsize=(10, 5))

#create a scatter plot using the 'plot' function of the DataFrame 'games'
#plotting 'NA_Sales' on the x-axis and 'JP_Sales' on the y-axis
games.plot(kind='scatter', x='NA_Sales', y='JP_Sales')

#set the title of the scatter plot to 'Video Games Sales in North America vs Japan'
plt.title('Video Games Sales in North America vs Japan')

#display the scatter plot
plt.show()
```

Generating Facet Grid

```python
#set the size of the figure for the plot (width: 10, height: 5)
plt.figure(figsize=(10, 5))

#create a FacetGrid using Seaborn (sns) library, using 'Genre' for color distinction and 'husl' palette for colors
#map a scatter plot onto the grid with 'Year' on the x-axis and 'Global_Sales' on the y-axis
#'add_legend()' adds a legend to the plot to indicate different genres
sns.FacetGrid(games, hue='Genre', palette='husl').map(plt.scatter, 'Year', 'Global_Sales').add_legend()

#display the plot
plt.show()
```

Generating Violin Plot

```python
#set the size of the figure for the plot (width: 10, height: 5)
plt.figure(figsize=(10, 5))

#create a violin plot using Seaborn (sns) library, mapping 'Genre' on the x-axis and 'Global_Sales' on the y-axis
#'palette' parameter sets the color palette to 'husl'
#'data=games' specifies the DataFrame to use for plotting
sns.violinplot(x='Genre', y='Global_Sales', palette='husl', data=games)

#set the title of the violin plot to describe the distribution of global video game sales based on genre
plt.title('Distribution of Global Video Game Sales Based on Genre')

#set the y-axis scale to logarithmic for better visualization of distribution across a wide range of values
plt.yscale('log')

#rotate x-axis labels (genres) by 45 degrees for better readability
plt.xticks(rotation=45)

#display the violin plot
plt.show()
```

Generating Box Plot

```python
#set the size of the figure for the plot (width: 10, height: 5)
plt.figure(figsize=(10, 5))

#create a box plot using Seaborn (sns) library, mapping 'Genre' on the x-axis and 'Global_Sales' on the y-axis
#'palette' parameter sets the color palette to 'husl'
#'data=games' specifies the DataFrame to use for plotting
sns.boxplot(x='Genre', y='Global_Sales', palette='husl', data=games)

#set the title of the box plot to describe the distribution of global video game sales based on genre
plt.title('Distribution of Global Video Game Sales Based on Genre')

#set the y-axis scale to logarithmic for better visualization of distribution across a wide range of values
plt.yscale('log')

#rotate x-axis labels (genres) by 45 degrees for better readability
plt.xticks(rotation=45)

#display the box plot
plt.show()
```

Generating KDE Plot

```python
#set the size of the figure for the plot (width: 10, height: 5)
plt.figure(figsize=(10, 5))

#create a FacetGrid using Seaborn (sns) library, using 'Genre' for color distinction and 'husl' palette for colors
#map a kernel density estimate plot onto the grid with 'Global_Sales' for each 'Genre'
#'add_legend()' adds a legend to the plot to indicate different genres
sns.FacetGrid(games, hue='Genre', palette='husl').map(sns.kdeplot, 'Global_Sales').add_legend()

#set the y-axis scale to logarithmic for better visualization of distribution across a wide range of values
plt.yscale('log')

#display the plot
plt.show()
```

Generating Pair Plot

```python
#create a pairplot using Seaborn (sns) library, displaying pairwise relationships between numerical columns in the 'games' DataFrame
#'hue' parameter uses 'Genre' for color distinction, and 'husl' palette specifies the color scheme
sns.pairplot(games, hue='Genre', palette='husl')

#display the pairplot
plt.show()
```
